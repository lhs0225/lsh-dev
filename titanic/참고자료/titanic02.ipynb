```python
'''
1. 데이터 불러오기
2. 데이터 전처리
    -데이터 확인하기
    -이상치 제거    #여기서는 생략합니다.
    -결측치 채우기
    -정규화,표준화
    -인코딩
3. 모델 만들기
4. 모델 학습하기
5. 모델 예측하기
6. 모델 평가
'''
```

```python
#1. 타이타닉 훈련 데이터 불러오기
train_df = pd.read_csv('train.csv')  #경로 지정 해주세요
train_df
```

```python
titanic_df=train_df.copy()  #원본데이터 카피 문제 생기면 여기서부터 다시 실행해보세요.
titanic_df.head()
```

```python
#2. 데이터 전처리
# 데이터 확인하기
titanic_df.info()
```

```python
'''
 0   PassengerId  891 non-null    int64     insert index
 1   Survived     891 non-null    int64     생존여부    0이면 사망, 1이면 생존
 2   Pclass       891 non-null    int64     티켓 등급   1,2,3
 3   Name         891 non-null    object    이름
 4   Sex          891 non-null    object    성별
 5   Age          714 non-null    float64   나이
 6   SibSp        891 non-null    int64     형제, 자매, 배우자의 합
 7   Parch        891 non-null    int64     부모, 자식의 합
 8   Ticket       891 non-null    object    티켓번호
 9   Fare         891 non-null    float64   비용
 10  Cabin        204 non-null    object    객실 번호
 11  Embarked     889 non-null    object    선착장 : C(Cherbourg), Q(Queenstown), S(Southampton)
 '''
```

```python
#결측치 확인 
titanic_df.isnull().sum() #isnull() isna() 똑같은거입니다. 
```

```python
#결측치 채우기
def fillna(df):
    df['Age'].fillna(df['Age'].mean(), inplace=True)  # mean을 median으로 바꾸면 중앙값으로 채울수 있습니다.
    #df['Cabin'].fillna()
    df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)  #mode는 최빈값(가장 많은 값)으로 구하는 메소드 입니다.
    df['Fare'].fillna(0, inplace=True)
    return df

#불필요한속성제거
def drop_features(df):
    df.drop(['PassengerId','Name','Ticket','Cabin'],axis=1,inplace=True)  #편의상 Cabin을 드랍에 추가했습니다.
    return df

#데이터 전처리 함수
def transform_features(df):
    df = fillna(df)
    df = drop_features(df)
    return df
```

```python
#데이터 전처리
transform_features(titanic_df)
```

```python
#결측치 확인 
titanic_df.isnull().sum() #isnull() isna() 똑같은거입니다. 
```

```python
#원-핫 인코딩
#Pclass는 타입이 int이기 때문에 원핫 인코딩하면 안먹습니다. 원한 다면 타입변경후 원핫 인코딩하셔야 나눌수 있습니다.

titanic_df['Pclass']=titanic_df['Pclass'].astype(str)
titanic_df=pd.get_dummies(titanic_df)
titanic_df.head()
```

```python
#변수와 값 나누기
feature=titanic_df.drop('Survived',axis=1)   #생존외 나머지
label=titanic_df['Survived']                 #생존
```

```python
from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score
```

```python
#하이퍼 파라미터 선언

seed=42 #랜덤 시드 설정

```

```python
#훈련, 학습 데이터 나누기
x_train,x_test,y_train,y_test=train_test_split(feature,label,test_size=0.25, random_state=seed)  
# x_train,y_train 로 모델을 학습
# x_test 로 예측
# y_test 로 정확도 측정 할겁니다.
```

```python
x_train.shape,y_train.shape ,x_test.shape,y_test.shape  #훈련,학습 데이터 크기 
```

```python
#모델 선언
model=DecisionTreeClassifier() # 모델 바꿔가면서 해보세요.
```

```python
#모델 학습
model.fit(x_train,y_train)
```

```python
#모델 예측
prediction=model.predict(x_test)
```

```python
#정확도 평가
accuracy_score(y_test,prediction)
```
